```python
df = pd.get_dummies(df, columns=["Product Category","Product Position","Seasonal", "brand","currency","terms","section","Promotion"])
```

```python
df.head()
```

```python
df = df.drop("name", axis=1, errors='ignore')
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, accuracy_score, confusion_matrix
```

```python
X= df.drop("price", axis=1)
y= df["price"]
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, random_state=42, test_size=42)
```

```python
models=[LinearRegression(), RandomForestRegressor()]
for model in models:
  model.fit(X_train, y_train)
  y_pred= model.predict(X_test)
  mse= mean_squared_error(y_test, y_pred)
  r2= r2_score(y_test, y_pred)
  mae= mean_absolute_error(y_test, y_pred)
  residuals=y_test-y_pred
  print(f"Model_name: {model.__class__.__name__}")
  print(f"MSE: {mse}")
  print(f"R2: {r2}")
  print(f"MAE: {mae}")
  sns.histplot(residuals, kde=True)
  plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV

# Adding Gradient Boosting Regressor to the models list
models.append(GradientBoostingRegressor())
param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5]
}

grid_search_rf = GridSearchCV(RandomForestRegressor(random_state=42), param_grid_rf, cv=3, scoring='r2', n_jobs=-1)

# Add the GridSearchCV object for tuning to a list if you want to tune multiple models
tuned_models = [grid_search_rf]

for model in models:
    if isinstance(model, (LinearRegression, GradientBoostingRegressor)): # Tune RandomForest separately
        model.fit(X_train, y_train)
        y_pred= model.predict(X_test)
        mse= mean_squared_error(y_test, y_pred)
        r2= r2_score(y_test, y_pred)
        mae= mean_absolute_error(y_test, y_pred)
        residuals=y_test-y_pred
        print(f"Model_name: {model.__class__.__name__}")
        print(f"MSE: {mse}")
        print(f"R2: {r2}")
        print(f"MAE: {mae}")
        sns.histplot(residuals, kde=True)
        plt.title(f"Residuals Distribution for {model.__class__.__name__}")
        plt.show()

for tuned_model in tuned_models:
    print(f"\nTuning model: {tuned_model.estimator.__class__.__name__}")
    tuned_model.fit(X_train, y_train)
    best_model = tuned_model.best_estimator_
    print(f"Best parameters found: {tuned_model.best_params_}")

    y_pred_tuned = best_model.predict(X_test)
    mse_tuned = mean_squared_error(y_test, y_pred_tuned)
    r2_tuned = r2_score(y_test, y_pred_tuned)
    mae_tuned = mean_absolute_error(y_test, y_pred_tuned)
    residuals_tuned = y_test - y_pred_tuned

    print(f"Tuned Model_name: {best_model.__class__.__name__}")
    print(f"MSE: {mse_tuned}")
    print(f"R2: {r2_tuned}")
    print(f"MAE: {mae_tuned}")
    sns.histplot(residuals_tuned, kde=True)
    plt.title(f"Residuals Distribution for Tuned {best_model.__class__.__name__}")
    plt.show()
```
